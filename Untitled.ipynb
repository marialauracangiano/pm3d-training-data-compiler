```python
import pandas as pd

# Load your data
master_df = pd.read_csv("data/processed/calibration_dataset.csv")
remove_df = pd.read_csv("B4I-Data_Report summary 12.9.25.csv")

# (Optional but recommended) align types
master_df["plot_id"] = master_df["plot_id"].astype(str)
remove_df["plot_id"] = remove_df["plot_id"].astype(str)

# Step 1: define statuses to remove
statuses_to_remove = ["remove", "used biomass, remove"]

# Step 2: filter plot_ids
plot_ids_to_remove = remove_df.loc[
    remove_df["status (pth 12.8.25)"].isin(statuses_to_remove),
    "plot_id"
]

# Step 3: drop from master
cleaned_df = master_df[~master_df["plot_id"].isin(plot_ids_to_remove)].reset_index(drop=True)
```

```python
print(len(master_df), "before")
print(len(cleaned_df), "after")
print(len(plot_ids_to_remove), "expected removed")
```
